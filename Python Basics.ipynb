```python
#Practical Question
'''
6. Write a function to demonstrate the behavior of mutable and immutable arguments.

Ans
'''
def modify_list(lst):
    lst.append(4)
    return lst

def modify_int(n):
    n += 1
    return n

# Mutable
my_list = [1, 2, 3]
print("Modified list:", modify_list(my_list))

# Immutable
num = 5
print("Modified number:", modify_int(num))
```
